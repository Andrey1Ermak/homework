```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore')
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(10000):
    r = 100
    x1.append(i)
    x2.append(-i)
    y1.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))

plt.figure(figsize = (10, 10))    
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.plot(x2, y1)
plt.grid(True)
plt.show()
```

```python
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(10000):
    a = 200
    b = 100
    x1.append(-i)
    x2.append(i)
    y1.append(np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))

plt.figure(figsize = (10, 5)) 
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.plot(x2, y1)
plt.grid(True)
plt.show()
```

```python
fig = figure(figsize = (10, 10))
ax = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
a = 3
b = 4
z1 = (x ** 2 / a ** 2) + (y ** 2 / b ** 2)
z2 = (x ** 2 / a ** 2) - (y ** 2 / b ** 2)
ax.plot_wireframe(x, y, z1, color = "red")
ax.plot_surface(x, y, z2)
show()
```
