```python
a = 0
b = 0
k = int(input('Введите количество экспериментов: '))

for i in range(k):
    np.random.randint(0, 36)
    if i % 2 == 0:
        a += 1
    elif i == 7:
        b += 1   
    else:
        pass

if a / k + b / k == (a + b) / k:
    print(f'Верятность выпадения четного - {a / k}\n'
          f'Верятность выпадения семерки - {b / k}\n'
          f'Верятность выпадения четного и семерки - {(a + b) / k}\n'
          'Теорема доказана')
```

```python
sum = 0
for i in range(10):
    i = np.random.randint(low = 0, high =10 , size = 1000)
    sum += i
```
