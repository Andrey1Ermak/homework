```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('fivethirtyeight')
```

```python
a = pd.read_csv('creditcard.csv')
```

```python
classes = a['Class'].value_counts()
classes
```

```python
classes.plot(kind='barh')
plt.show()
```

```python
classes.plot(kind='barh', logx=True)
plt.show()
```

```python
legal= a.loc[a['Class'] == 0, 'V1']
illegal = a.loc[a['Class'] == 1, 'V1']
```

```python
plt.hist([illegal, legal], bins=20, density=True, color=['red','grey'], alpha=0.5)
label_font = {
    "fontsize": 16,
    "fontweight":"bold",
    "family": "Arial",
}
plt.xlabel('V1', fontdict=label_font)
plt.legend(labels=['Class1', 'Class0'], frameon=False, loc="lower left")

plt.show()
```
