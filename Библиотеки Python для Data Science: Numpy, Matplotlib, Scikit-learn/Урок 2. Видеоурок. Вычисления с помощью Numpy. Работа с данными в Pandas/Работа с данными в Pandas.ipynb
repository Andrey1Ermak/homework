```python
import pandas as pd
import numpy as np
```

```python
a = {
    "author_id": [1, 2, 3],
    "author_name": ['Тургенев', 'Чехов', 'Островский']
}
authors = pd.DataFrame(a)
authors
```

```python
b = {
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 
                   'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price": [450, 300, 350, 500, 450, 370, 290]
}
book = pd.DataFrame(b)
book
```

```python
authors_price = pd.merge(authors, book, on='author_id', how='outer')
authors_price
```

```python
authors_price.sort_values(by="price", ascending=False, inplace=True)
authors_price.set_index("author_id", inplace=True)
top5 = authors_price.head()
top5
```

```python
groupby = authors_price.groupby("author_name")
groupby.agg({'price': ['min', 'max', 'mean']}) 
```

```python
authors_price["cover"] = ['твердая', 'мягкая', 'мягкая', 'твердая', 
                          'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
new = pd.pivot_table(authors_price,
                     index=["author_name"],
                     values=["price"],
                     columns=["cover"],
                     aggfunc=[np.sum])
book_info = new.fillna(0)
book_info
```

```python
book_info.to_pickle("book_info.pkl")
```
