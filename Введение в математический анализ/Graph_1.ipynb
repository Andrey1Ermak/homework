```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.linspace(-50, 50, 100)
plt.plot(x, x**2 + 4*x + 3)
plt.grid()
plt.show()

x = np.linspace(-5, 5, 100)
plt.plot(x, -2 * np.sin(3*x))
plt.grid()
plt.show()

x = np.linspace(-10, 10, 100)
y = []
for i in x:
    a = abs(i % int(i)) - 1/2
    y.append(a)
plt.plot(x, y)
plt.grid()
plt.show()
```
