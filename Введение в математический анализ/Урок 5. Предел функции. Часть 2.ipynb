```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
np.pi
```

```python
one = np.linspace(-3 , 3, 1000)

x1 = np.array([x for x in one if (abs(x) <= 1)])
plt.plot(x1, (np.cos(np.pi * x1 / 2)))

x2 = np.array([x for x in one if x >= 1])
sns.lineplot(x2, abs(x2 - 1))

x3 = np.array([x for x in one if x <= -1])
sns.lineplot(x3, abs(x3 - 1))

plt.grid()
plt.show()
```
